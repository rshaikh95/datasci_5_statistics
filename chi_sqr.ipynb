```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python

df = pd.read_csv('/content/healthcare-dataset-stroke-data.csv') # noqa
df
```

```python
df['smoking_status'].value_counts()

```

```python
df['Residence_type'].value_counts()
```

### **There's an association between residence type and smoking status. I believe that there will be more smokers in urban environments.**

Null Hypothesis:  H0: There is no relationship between residence type and smoking status.

Alternative Hypothesis:  Ha: There is a relationship between residence type and smoking status.

```python
contingency_table = pd.crosstab(df['smoking_status'], df['Residence_type'])
print(contingency_table)
```

```python
chi2, p, degreesoffreedom, expected = chi2_contingency(contingency_table)
print(f"Chi2 value: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {degreesoffreedom}")
print(f"Expected: {expected}")
```

## There isn't a definitive relationship between residence type and smoking status therefore we accept null hypothesis.

 For this Chi-Square test, I obtained a dataset from Kaggle called Stroke Prediction Dataset (link in readme of github repo). I used the columns for smoking status and residence type to see if there is any relationship between smoking status and thier residence. The options for smoking status was formerly smoked, never smoked, smokes and unknown. I converted the unknown into tried smoking once to fill in missing data on the excel sheet. Finally I obtained the values for chi square, probability, degrees of freedom and expected. I analyzed the result and came up with the following. **Living in a rural or urban area does not affect smoking status of a person.**
